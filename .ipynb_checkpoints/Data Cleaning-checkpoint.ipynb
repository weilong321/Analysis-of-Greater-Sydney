# Task 1

```python
# import libaries

import pandas as pd
import geopandas as gpd
from shapely.geometry import Point, Polygon, MultiPolygon
from shapely import wkt
from geoalchemy2 import Geometry, WKTElement
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# functions

srid = 4326

def drop_cols(df): 
    to_drop = []
    for col in df.columns:
        if len(df[col].unique()) == 1:
            to_drop.append(col)
    return to_drop

def create_wkt_element(geom, srid):
    if geom.geom_type == 'Polygon':
        geom = MultiPolygon([geom])
    return WKTElement(geom.wkt, srid)
```

```python
# import datasets

businesses = pd.read_csv("datasets\Businesses.csv")
income = pd.read_csv("datasets\Income.csv")
polling_places_2019 = pd.read_csv("datasets\PollingPlaces2019.csv")
population = pd.read_csv("datasets\Population.csv")
stops = pd.read_csv("datasets\Stops.txt")
catchments_secondary = gpd.read_file("datasets\Catchments\catchments\catchments_secondary.shp")
catchments_primary = gpd.read_file("datasets\Catchments\catchments\catchments_primary.shp")
catchments_future = gpd.read_file("datasets\Catchments\catchments\catchments_future.shp")
sa2_regions = gpd.read_file("datasets\SA2_2021_AUST_SHP_GDA2020/SA2_2021_AUST_GDA2020.shp")
crime = gpd.read_file("datasets\crime\StealFromVehicle_JanToDec2021.shp")
births = gpd.read_file("datasets\\births.json")
```

```python
cleaned_sa2 = sa2_regions.copy()
cleaned_sa2 = cleaned_sa2.query('GCC_NAME21 == "Greater Sydney"')
to_drop_sa2 = drop_cols(cleaned_sa2)
to_drop_sa2.append("LOCI_URI21")
cleaned_sa2.drop(to_drop_sa2, axis=1, inplace=True)
cleaned_sa2['geom'] = cleaned_sa2['geometry'].apply(lambda x: create_wkt_element(geom=x,srid=srid))
cleaned_sa2 = cleaned_sa2.drop(columns="geometry")
sa2_codes = cleaned_sa2.SA2_CODE21.reset_index(drop=True)
```

```python
cleaned_crime = crime.copy()
drop = ['ORIG_FID']
cleaned_crime.drop(drop, inplace=True, axis=1)
cleaned_crime.columns = cleaned_crime.columns.str.lower()
cleaned_crime['geom'] = cleaned_crime['geometry'].apply(lambda x: create_wkt_element(geom=x,srid=srid))
cleaned_crime = cleaned_crime.drop(columns="geometry")
```

```python
cleaned_births = births.copy()
cleaned_births = cleaned_births[['sa2_code', 'sa2_name', '_2020_births_no', '_2020_total_fertility_rate_rate', 'geometry']]
cleaned_births['sa2_code'] = cleaned_births['sa2_code'].astype(str)
cleaned_births = cleaned_births[cleaned_births['sa2_code'].isin(sa2_codes)]
cleaned_births['_2020_total_fertility_rate_rate'] = cleaned_births['_2020_total_fertility_rate_rate'].replace('np', 0)
cleaned_births = cleaned_births[cleaned_births['_2020_births_no'] != 0] 
cleaned_births = cleaned_births[cleaned_births['_2020_total_fertility_rate_rate'] != 0]
cleaned_births['geom'] = cleaned_births['geometry'].apply(lambda x: create_wkt_element(geom=x,srid=srid))
cleaned_births = cleaned_births.drop(columns="geometry")
cleaned_births.reset_index(drop=False, inplace=True)
cleaned_births.rename(columns={"index": "id", "_2020_births_no": "births_no", "_2020_total_fertility_rate_rate": "total_fertility_rate"}, inplace = True)
```

```python
cleaned_businesses = businesses.copy()
cleaned_businesses = cleaned_businesses.drop(cleaned_businesses.columns[4:10], axis=1)
cleaned_businesses['sa2_code'] = cleaned_businesses['sa2_code'].astype(str)
cleaned_businesses = cleaned_businesses[cleaned_businesses['sa2_code'].isin(sa2_codes)]
cleaned_businesses.reset_index(drop=False, inplace=True)
cleaned_businesses.rename(columns={'index': 'business_id'}, inplace=True)
```

```python
cleaned_income = income.copy()
cleaned_income['sa2_code'] = cleaned_income['sa2_code'].astype(str)
cleaned_income = cleaned_income[cleaned_income['sa2_code'].isin(sa2_codes)]
cleaned_income = cleaned_income[cleaned_income['median_income'] != "np"]
cleaned_income = cleaned_income[cleaned_income['mean_income'] != "np"]
cleaned_income.reset_index(drop=False, inplace=True)
cleaned_income.rename(columns={'index': 'income_id'}, inplace=True)
```

```python
cleaned_polling = polling_places_2019.copy()
to_drop_polling = drop_cols(cleaned_polling)
cleaned_polling['FID'] = cleaned_polling['FID'].str.replace("aec_federal_election_polling_places_2019.fid-492f303f_1873b973d45_-", "", regex=False)
cleaned_polling.drop(to_drop_polling, axis=1, inplace=True)
cleaned_polling['FID'] = cleaned_polling['FID'].astype(str)
cleaned_polling['temp_geom'] = gpd.points_from_xy(cleaned_polling.longitude, cleaned_polling.latitude)
cleaned_polling['geom'] = cleaned_polling['temp_geom'].apply(lambda x: WKTElement(x.wkt, srid=srid))
cleaned_polling = cleaned_polling.drop(['the_geom', 'temp_geom', 'latitude', 'longitude'], axis=1)
```

```python
cleaned_population = pd.DataFrame()
cleaned_population['sa2_code'] = population['sa2_code']
cleaned_population['sa2_name'] = population['sa2_name']
cleaned_population['young_people'] = population['0-4_people'] + population['5-9_people'] + population['10-14_people'] + population['15-19_people']
cleaned_population['total_people'] = population['total_people']
cleaned_population['sa2_code'] = cleaned_population['sa2_code'].astype(str)
cleaned_population = cleaned_population[cleaned_population['sa2_code'].isin(sa2_codes)]
cleaned_population.reset_index(drop=False, inplace=True)
cleaned_population.rename(columns={'index': 'population_id'}, inplace=True)
```

```python
cleaned_stops = stops.copy()
cleaned_stops['geom'] = gpd.points_from_xy(cleaned_stops.stop_lon, cleaned_stops.stop_lat)
cleaned_stops['geom'] = cleaned_stops['geom'].apply(lambda x: WKTElement(x.wkt, srid=srid))
cleaned_stops = cleaned_stops.drop(columns=['stop_lon', 'stop_lat'])
```

```python
cleaned_future = catchments_future.copy()
cleaned_primary = catchments_primary.copy()
cleaned_secondary = catchments_secondary.copy()

cleaned_schools = pd.concat([cleaned_future, cleaned_primary, cleaned_secondary], axis=0)
cleaned_schools = cleaned_schools.drop(cleaned_schools.columns[3:17], axis=1)
cleaned_schools = cleaned_schools.drop(cleaned_schools.columns[4], axis=1)

cleaned_schools['geom'] = cleaned_schools['geometry'].apply(lambda x: create_wkt_element(geom=x,srid=srid))
cleaned_schools = cleaned_schools.drop(columns='geometry')

cleaned_schools = cleaned_schools.drop_duplicates(subset='USE_ID', keep='first')
```

```python
engine = create_engine('postgresql+psycopg2://postgres:password@localhost/test', echo=False)

conn = psycopg2.connect(
    host="localhost",
    database="test",
    user="postgres",
    password="password"
)

def query(conn, file, args=None, df=True):
    with open(file, 'r') as file:
        sqlcmd = file.read()    
    result = pd.DataFrame() if df else None
    try:
        if df:
            result = pd.read_sql_query(sqlcmd, conn, params=args)
        else:
            result = conn.execute(sqlcmd, args).fetchall()
            result = result[0] if len(result) == 1 else result
    except Exception as e:
        print("Error encountered: ", e, sep='\n')
    return result
```

```python
cleaned_sa2.to_sql('sa2_regions', engine, schema='public', if_exists='append', index=False, dtype={'geom': Geometry('MULTIPOLYGON', srid)})
cleaned_births.to_sql("births", engine, schema='public', if_exists='append', index=False, dtype={'geom': Geometry('MULTIPOLYGON', srid)})
cleaned_crime.to_sql("crime", engine, schema='public', if_exists='append', index=False, dtype={'geom': Geometry('MULTIPOLYGON', srid)})
cleaned_businesses.to_sql('businesses', engine, schema='public', if_exists='append', index=False)
cleaned_income.to_sql('income', engine, schema='public', if_exists='append', index=False)
cleaned_population.to_sql('population', engine, schema='public', if_exists='append', index=False)
cleaned_polling.to_sql('polling_places', engine, schema='public', if_exists='append', index=False, dtype={'geom': Geometry('POINT', srid)})
cleaned_stops.to_sql('stops', engine, schema='public', if_exists='append', index=False, dtype={'geom': Geometry('POINT', srid)})
cleaned_schools.to_sql('schools', engine, schema='public', if_exists='append', index=False, dtype={'geom': Geometry('MULTIPOLYGON', srid)})
```

```python
df = query(conn, 'map_overlay_plot_2.sql')

temp = sa2_regions.copy()
temp = temp.query('GCC_NAME21 == "Greater Sydney"')
temp['SA2_CODE21'] = temp['SA2_CODE21'].astype('int64')
cleaned_sa2_subset = temp[['SA2_CODE21', 'geometry']]

merged_df = pd.merge(df, cleaned_sa2_subset, left_on='sa2_code', right_on='SA2_CODE21', how='inner')
merged_df = merged_df.drop(columns = 'SA2_CODE21')

ax = sa2_regions.plot(color='lightgray', edgecolor='white')
ax.set_xlim(149.95, 151.65)
ax.set_ylim(-34.4, -32.95)
ax.set_title('Task 2')
gdf = gpd.GeoDataFrame(merged_df, geometry='geometry')
gdf.plot(column='sigmoid', legend=True, ax=ax)
```
